### Three-level Lambda Atom Coupled to a Waveguide


```python
from cmath import exp
from math import pi
from matplotlib import pyplot as plt
from matplotlib import rc
import numpy as np
import time
#rc('font',**{'family':'serif','serif':['Palatino']})
plt.rcParams.update({'font.size': 13})
rc('text', usetex=True)
```

Define transmission and reflection in terms of energy (not frequency - E is conserved but k is not)

```python
def trans1(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |1>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return 1 - (g1**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref1(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |1>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return - (g1**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def trans3(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |3>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref3(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |3>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def trans1p(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |1>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref1p(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |1>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def trans3p(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |3>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return 1 - (g3**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref3p(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |3>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return - (g3**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def tmatrix(E2,E,vg,g1,g3):
    """
    transfer matrix of three level atom
    """
    t1 = trans1(E2,E,vg,g1,g3)
    r1 = ref1(E2,E,vg,g1,g3)
    t3 = trans3(E2,E,vg,g1,g3)
    r3 = ref3(E2,E,vg,g1,g3)
    t1p = trans1p(E2,E,vg,g1,g3)
    r1p = ref1p(E2,E,vg,g1,g3)
    t3p = trans3p(E2,E,vg,g1,g3)
    r3p = ref3p(E2,E,vg,g1,g3)
    # matrix elements
    denominator = (t1p*t3 - t1*t3p)
    if (t1p*t3 - t1*t3p) == 0:
        denominator = 0.00000001 #avoid divide by zero error
    M11 = -( (-r1p*r3*t1 + r1*r3*t1p + r1*r1p*t3 - t1*t1p*t3 - r1**2*t3p + t1**2*t3p)/denominator)
    M12 = -( (-r1p*t3 + r1*t3p)/denominator)
    M13 = -( (-r1p*r3p*t1 + r1*r3p*t1p + r1p**2*t3 - t1p**2*t3 - r1*r1p*t3p + t1*t1p*t3p)/denominator)
    M14 = -( (r1p*t1 - r1*t1p)/denominator)
    M21 = -( (r3*t1p - r1*t3p)/denominator)
    M22 = -( t3p/denominator)
    M23 = -( (r3p*t1p - r1p*t3p)/denominator)
    M24 =  t1p/denominator
    M31 = -( (-r3*r3p*t1 + r3**2*t1p + r1*r3p*t3 - t1p*t3**2 - r1*r3*t3p + t1*t3*t3p)/denominator)
    M32 = -( (-r3p*t3 + r3*t3p)/denominator)
    M33 = -( (-r3p**2*t1 + r3*r3p*t1p + r1p*r3p*t3 - r1p*r3*t3p - t1p*t3*t3p + t1*t3p**2)/denominator)
    M34 = -( (r3p*t1 - r3*t1p)/denominator)
    M41 = -((-r3*t1 + r1*t3)/denominator)
    M42 = +( t3/denominator)
    M43 = -( (-r3p*t1 + r1p*t3)/denominator)
    M44 = -( t1/denominator)
    return np.array([[M11,M12,M13,M14],[M21,M22,M23,M24], [M31,M32,M33,M34],[M41,M42,M43,M44]])
```

Plot t and r analytically

```python
# Define variables
E2=20
E1=0
E3=10
vg = 1
g1 = 1
g3 = 1
dl = 3.4 * 10**(-1)

# Define plot range
E_min = 0
E_max = 30

steps = 1000
E_array = np.linspace(E_min, E_max, steps)
    
#plot from analytical
plt.plot(E_array,[abs(trans1(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|t_1|^2$')
plt.plot(E_array,[abs(ref1(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|r_1|^2$')
plt.plot(E_array,[abs(trans3(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|t_3|^2$')
plt.plot(E_array,[abs(ref3(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|r_3|^2$')
plt.plot(E_array,[abs(trans1p(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|t_1\prime|^2$')
plt.plot(E_array,[abs(ref1p(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|r_1\prime|^2$')
plt.plot(E_array,[abs(trans3p(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|t_3\prime|^2$')
plt.plot(E_array,[abs(ref3p(E2,E,vg,g1,g3))**2 for E in E_array], label= '$|r_3\prime|^2$')
plt.xlabel("$E$")
plt.legend()
plt.show()
```

Plot dispersion relation

```python
# Define variables
E2=20
E1=0
E3=10
vg = 1
g1 = 1
g3 = 1
llambda = 2*pi*vg/(E2-E1)
d = 3.4 * 10**(-1) *llambda

# Define plot range
E_min = 0
E_max = 42
steps = 4000
E_array = np.linspace(E_min, E_max, steps)

eig_array1 = []
eig_array2 = []
eig_array3 = []
eig_array4 = []

for E in E_array:
    Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
    Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
    eig = np.linalg.eigvals(Q)
    #print(np.around(eig,2)) #-> check determinant is 1
    eig_array1.append(eig[0])
    eig_array2.append(eig[1])
    eig_array3.append(eig[2])
    eig_array4.append(eig[3])
    
plt.scatter(np.log(eig_array1)/1j,E_array,s=0.2,c='k')
plt.scatter(np.log(eig_array2)/1j,E_array,s=0.2,c='k')
plt.scatter(np.log(eig_array3)/1j,E_array,s=0.2,c='k')
plt.scatter(np.log(eig_array4)/1j,E_array,s=0.2,c='k')
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.title("$E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.xlim([-0,pi])
plt.show()
```

```python
# these all assume right going photon and no left going input states
# input states |psi>_in = r1m|k,1>_R + r3m|k-Delta,3>_R + l1p|k,1>_L + l3p|k-Delta,3>_L
# = AC|k,1>_R and BCexp(1j*(phi_a+phi_p))|k-Delta,3>_R
# outgoing states |psi>_out = r1p|k,1>_R + r3p|k-Delta,3>_R + l1m|k,1>_L + l3m|k-Delta,3>_L

def output_states(matrix,A,C,B,D,phi_a,phi_p,l1p,l3p,r1m,r3m):
    """
    takes in a 4 by 4 transfer matrix, input variables
    outputs the amplitude R1+
    """
    m11, m12, m13, m14 = matrix[0]
    m21, m22, m23, m24 = matrix[1]
    m31, m32, m33, m34 = matrix[2]
    m41, m42, m43, m44 = matrix[3]

    r1p = -((l3p*(m14*m22 - m12*m24))/(m24*m42 - m22*m44)) - (
      l1p*(-m14*m42 + m12*m44))/(
      m24*m42 - 
       m22*m44) - ((-m14*m22*m41 + m12*m24*m41 + m14*m21*m42 - 
           m11*m24*m42 - m12*m21*m44 + m11*m22*m44)*r1m)/(m24*m42 - 
         m22*m44) - ((m14*m23*m42 - m13*m24*m42 - m14*m22*m43 + 
           m12*m24*m43 + m13*m22*m44 - m12*m23*m44)*r3m)/(m24*m42 - 
         m22*m44)

    r3p = -((l3p*(-m24*m32 + m22*m34))/(m24*m42 - m22*m44)) - (
      l1p*(-m34*m42 + m32*m44))/(
      m24*m42 - 
       m22*m44) - ((m24*m32*m41 - m22*m34*m41 - m24*m31*m42 + 
           m21*m34*m42 + m22*m31*m44 - m21*m32*m44)*r1m)/(m24*m42 - 
         m22*m44) - ((-m24*m33*m42 + m23*m34*m42 + m24*m32*m43 - 
           m22*m34*m43 - m23*m32*m44 + m22*m33*m44)*r3m)/(m24*m42 - 
         m22*m44)

    l1m = (l3p*m24)/(m24*m42 - m22*m44) - (l1p*m44)/(
      m24*m42 - m22*m44) - ((m24*m41 - m21*m44)*r1m)/(
      m24*m42 - m22*m44) - ((m24*m43 - m23*m44)*r3m)/(m24*m42 - m22*m44)

    l3m = -((l3p*m22)/(m24*m42 - m22*m44)) + (l1p*m42)/(
      m24*m42 - m22*m44) - ((-m22*m41 + m21*m42)*r1m)/(
      m24*m42 - m22*m44) - ((m23*m42 - m22*m43)*r3m)/(m24*m42 - m22*m44)
    
    if r1p > 1:
        r1p = 0
    if r3p > 1:
        r3p = 0
    if l1m > 1:
        l1m = 0
    if l3m > 1:
        l3m = 0

    
    return r1p, r3p, l1m, l3m
```

```python
# plot as a function of phi_a and phi_p 
n_atoms = 10
E2=20
E1=0
E3=10
vg = 1
g1 = 1
g3 = 1
llambda = 2*pi*vg/(E2-E1)
d = 3.4 * 10**(-1) *llambda
phi_a = 0

A, B, C, D = 1/np.sqrt(2), 1/np.sqrt(2), 1/np.sqrt(2), 1/np.sqrt(2)

phi_pp = np.linspace(-pi, pi, 100)
E_array = np.linspace(0, 40, 100)

#Energy subspace A
r1p_arrayA = np.zeros((100,100))
r3p_arrayA = np.zeros((100,100))
l1m_arrayA = np.zeros((100,100))
l3m_arrayA = np.zeros((100,100))
i, j = 0, 0
for E in E_array:
    j = 0
    i += 1
    for phi_p in phi_pp:
        l1p = 0
        l3p = 0
        r1m = B*C*exp(1j*phi_a)
        r3m = 0
        Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
        Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
        totaltmatrix = np.linalg.matrix_power(Q, n_atoms)
        r1p, r3p, l1m, l3m = output_states(totaltmatrix,A,C,B,D,phi_a,phi_p,l1p,l3p,r1m,r3m)
        r1p_arrayA[i-1,j-1] = r1p
        r3p_arrayA[i-1,j-1] = r3p
        l1m_arrayA[i-1,j-1] = l1m
        l3m_arrayA[i-1,j-1] = l3m
        j += 1
        
#Energy subspace B
r1p_arrayB = np.zeros((100,100))
r3p_arrayB = np.zeros((100,100))
l1m_arrayB = np.zeros((100,100))
l3m_arrayB = np.zeros((100,100))
i, j = 0, 0
for E in E_array:
    j = 0
    i += 1
    for phi_p in phi_pp:
        l1p = 0
        l3p = 0
        r1m = A*C
        r3m = B*D*exp(1j*(phi_a+phi_p))
        Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
        Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
        totaltmatrix = np.linalg.matrix_power(Q, n_atoms)
        r1p, r3p, l1m, l3m = output_states(totaltmatrix,A,C,B,D,phi_a,phi_p,l1p,l3p,r1m,r3m)
        r1p_arrayB[i-1,j-1] = r1p
        r3p_arrayB[i-1,j-1] = r3p
        l1m_arrayB[i-1,j-1] = l1m
        l3m_arrayB[i-1,j-1] = l3m
        j += 1
        
#Energy subspace C
r1p_arrayC = np.zeros((100,100))
r3p_arrayC = np.zeros((100,100))
l1m_arrayC = np.zeros((100,100))
l3m_arrayC = np.zeros((100,100))
i, j = 0, 0
for E in E_array:
    j = 0
    i += 1
    for phi_p in phi_pp:
        l1p = 0
        l3p = 0
        r1m = 0
        r3m = A*D*exp(1j*phi_a)
        Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
        Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
        totaltmatrix = np.linalg.matrix_power(Q, n_atoms)
        r1p, r3p, l1m, l3m = output_states(totaltmatrix,A,C,B,D,phi_a,phi_p,l1p,l3p,r1m,r3m)
        r1p_arrayC[i-1,j-1] = r1p
        r3p_arrayC[i-1,j-1] = r3p
        l1m_arrayC[i-1,j-1] = l1m
        l3m_arrayC[i-1,j-1] = l3m
        j += 1

#Energy A
plt.imshow([abs(x)**2 for x in r1p_arrayA],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy A, R1+ is $|k-\Delta,1>_R$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in r3p_arrayA],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy A, R3+ is $|k-2\Delta,3>_R$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in l1m_arrayA],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy A, L1- is $|k-\Delta,1>_L$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in l3m_arrayA],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy A, L3- is $|k-2\Delta,3>_L$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()

#Energy B
plt.imshow([abs(x)**2 for x in r1p_arrayB],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy B, R1+ is $|k,1>_R$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in r3p_arrayB],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy B, R3+ is $|k-\Delta,3>_R$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in l1m_arrayB],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy B, L1- is $|k,1>_L$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in l3m_arrayB],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy B, L3- is $|k-\Delta,3>_L$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()



#Energy C
plt.imshow([abs(x)**2 for x in r1p_arrayC],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy C, R1+ is $|k+\Delta,1>_R$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in r3p_arrayC],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy C, R3+ is $|k,3>_R$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in l1m_arrayC],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy C, L1- is $|k+\Delta,1>_L$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()
plt.imshow([abs(x)**2 for x in l3m_arrayC],extent=[-pi,pi,E_min,E_max],aspect="auto",origin='lower')
plt.xlabel("$\phi_p$")
plt.ylabel("$E$")
plt.title("Energy C, L3- is $|k,3>_L$\n $N_{atoms}=$"+str(n_atoms)+"$, \phi_a=$"+str(phi_a)+"$, E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.colorbar()
plt.show()

#sum  DO LATER
```
