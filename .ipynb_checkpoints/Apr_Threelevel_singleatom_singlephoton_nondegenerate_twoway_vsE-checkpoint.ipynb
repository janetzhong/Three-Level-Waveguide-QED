Define transmission and reflection in terms of energy (not frequency - E is conserved but k is not)

```python
def trans1(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |1>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return 1 - (g1**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref1(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |1>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return - (g1**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def trans3(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |3>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref3(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |3>, with atom initially in |1>
    """
    # E is total energy (E=E1+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def trans1p(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |1>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref1p(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |1>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return - (g1*g3)/(1j*(E2-E)*vg + g1**2 + g3**2)

def trans3p(E2,E,vg,g1,g3):
    """
    transmission of three-level atom through atom |3>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return 1 - (g3**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def ref3p(E2,E,vg,g1,g3):
    """
    reflection of three-level atom through atom |3>, with atom initially in |3>
    """
    # E is total energy (E=E3+Ephoton)
    return - (g3**2)/(1j*(E2-E)*vg + g1**2 + g3**2)

def tmatrix(E2,E,vg,g1,g3):
    """
    transfer matrix of three level atom
    """
    t1 = trans1(E2,E,vg,g1,g3)
    r1 = ref1(E2,E,vg,g1,g3)
    t3 = trans3(E2,E,vg,g1,g3)
    r3 = ref3(E2,E,vg,g1,g3)
    t1p = trans1p(E2,E,vg,g1,g3)
    r1p = ref1p(E2,E,vg,g1,g3)
    t3p = trans3p(E2,E,vg,g1,g3)
    r3p = ref3p(E2,E,vg,g1,g3)
    # matrix elements
    denominator = (t1p*t3 - t1*t3p)
    if (t1p*t3 - t1*t3p) == 0:
        nanmatrix = np.empty((4,4,))
        nanmatrix = np.nan
        return nanmatrix
        #denominator = 0.00000001 #avoid divide by zero error
    else:
        M11 = -( (-r1p*r3*t1 + r1*r3*t1p + r1*r1p*t3 - t1*t1p*t3 - r1**2*t3p + t1**2*t3p)/denominator)
        M12 = -( (-r1p*t3 + r1*t3p)/denominator)
        M13 = -( (-r1p*r3p*t1 + r1*r3p*t1p + r1p**2*t3 - t1p**2*t3 - r1*r1p*t3p + t1*t1p*t3p)/denominator)
        M14 = -( (r1p*t1 - r1*t1p)/denominator)
        M21 = -( (r3*t1p - r1*t3p)/denominator)
        M22 = -( t3p/denominator)
        M23 = -( (r3p*t1p - r1p*t3p)/denominator)
        M24 =  t1p/denominator
        M31 = -( (-r3*r3p*t1 + r3**2*t1p + r1*r3p*t3 - t1p*t3**2 - r1*r3*t3p + t1*t3*t3p)/denominator)
        M32 = -( (-r3p*t3 + r3*t3p)/denominator)
        M33 = -( (-r3p**2*t1 + r3*r3p*t1p + r1p*r3p*t3 - r1p*r3*t3p - t1p*t3*t3p + t1*t3p**2)/denominator)
        M34 = -( (r3p*t1 - r3*t1p)/denominator)
        M41 = -((-r3*t1 + r1*t3)/denominator)
        M42 = +( t3/denominator)
        M43 = -( (-r3p*t1 + r1p*t3)/denominator)
        M44 = -( t1/denominator)
        return np.array([[M11,M12,M13,M14],[M21,M22,M23,M24], [M31,M32,M33,M34],[M41,M42,M43,M44]])
```

Plot dispersion relation

```python
# Define variables
E2=20
E1=0
E3=0

# band diagram uncoupled atom and photon modes
vg = 1
g1 = 0
g3 = 0
llambda = 2*pi*vg/(E2-E1)
d = 3.4 * 10**(-1) *llambda
phi_a = 0

E_min = 0
E_max = 50
steps = 4000
E_array = np.linspace(E_min, E_max, steps)

eig_array1 = []
eig_array2 = []
eig_array3 = []
eig_array4 = []

for E in E_array:
    Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
    Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
    eig = np.linalg.eigvals(Q)
    #print(np.around(eig,2)) #-> check determinant is 1
    eig_array1.append(eig[0])
    eig_array2.append(eig[1])
    eig_array3.append(eig[2])
    eig_array4.append(eig[3])

# photon |1> , |3>
eig_array_concat = np.concatenate([eig_array1,eig_array2,eig_array3,eig_array4])
E_array_concat = np.concatenate([E_array,E_array,E_array,E_array])
plt.scatter(np.log(eig_array_concat)/1j,E_array_concat,s=0.5,c='r',zorder = 2)
# atom |2>
plt.axhline(y=E2, linewidth=3,c="r",zorder = 1)
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.title("$E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.xlim([-pi,pi])


# band diagrom coupled
g1 = 1
g3 = 1

# Define plot range
eig_array1 = []
eig_array2 = []
eig_array3 = []
eig_array4 = []

for E in E_array:
    Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
    Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
    eig = np.linalg.eigvals(Q)
    #print(np.around(eig,2)) #-> check determinant is 1
    eig_array1.append(eig[0])
    eig_array2.append(eig[1])
    eig_array3.append(eig[2])
    eig_array4.append(eig[3])
    
eig_array_concat = np.concatenate([eig_array1,eig_array2,eig_array3,eig_array4])
E_array_concat = np.concatenate([E_array,E_array,E_array,E_array])
plt.scatter(np.log(eig_array_concat)/1j,E_array_concat,s=2,c='k',zorder = 3)
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.title("$E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3)+\
         "\n $g_1=$"+str(g1)+", $g_3 =$"+str(g3))
plt.xlim([0,pi])
#plt.ylim([0,40])
plt.show()
```

Phase of eigenvectors in band diagram

```python
# Define variables
E2=20
E1=0
E3=0
vg = 1
g1 = 1
g3 = 1
llambda = 2*pi*vg/(E2-E1)
d = 3.4 * 10**(-1) *llambda

# Define plot range
E_min = 0
E_max = 60
steps = 4000
E_array = np.linspace(E_min, E_max, steps)

eig_array1 = []
eig_array2 = []
eig_array3 = []
eig_array4 = []
eigvec_array1 = []
eigvec_array2 = []
eigvec_array3 = []
eigvec_array4 = []

for E in E_array:
    Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
    Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
    try: 
        eig_val, eig_vec = np.linalg.eig(Q)
        #print(np.around(eig,2)) #-> check determinant is 1
        #print('eig 1: '+str(round(eig[0],2))+'eig 2: '+str(round(eig[1],2))+'eig 3: '+str(round(eig[2],2))+'eig 4: '+str(round(eig[3],2)))
        eig_array1.append(eig_val[0])
        eig_array2.append(eig_val[1])
        eig_array3.append(eig_val[2])
        eig_array4.append(eig_val[3])
        eigvec_array1.append(eig_vec[:,0])
        eigvec_array2.append(eig_vec[:,1])
        eigvec_array3.append(eig_vec[:,2])
        eigvec_array4.append(eig_vec[:,3])
    except:
        # fails if tmatrix is nans, which happens if transfer matrix element denominator is 0
#         nanarray = np.empty((1,4,))[0]
#         nanarray[:] = np.nan # 1x4 array of NaN
#         eig_val = nanarray
#         eig_vec = np.array([nanarray,nanarray,nanarray,nanarray])
        eig_array1.append(None)
        eig_array2.append(None)
        eig_array3.append(None)
        eig_array4.append(None)
        eigvec_array1.append(None)
        eigvec_array2.append(None)
        eigvec_array3.append(None)
        eigvec_array4.append(None)
        
# Plot dispersion relation while masking the NaN values
series = np.array(eig_array1).astype(np.complex)
mask = np.isfinite(series)
eig_array_mask1 = np.array(eig_array1)[mask].astype(complex)
eig_array_mask2 = np.array(eig_array2)[mask].astype(complex)
eig_array_mask3 = np.array(eig_array3)[mask].astype(complex)
eig_array_mask4 = np.array(eig_array4)[mask].astype(complex)
eigvec_array_mask1 = np.array(eigvec_array1)[mask]
eigvec_array_mask2 = np.array(eigvec_array2)[mask]
eigvec_array_mask3 = np.array(eigvec_array3)[mask]
eigvec_array_mask4 = np.array(eigvec_array4)[mask]
E_array_mask = E_array[mask]

plt.scatter(np.log(eig_array_mask1)/1j,E_array_mask,s=0.2,c='k')
plt.scatter(np.log(eig_array_mask2)/1j,E_array_mask,s=0.2,c='k')
plt.scatter(np.log(eig_array_mask3)/1j,E_array_mask,s=0.2,c='k')
plt.scatter(np.log(eig_array_mask4)/1j,E_array_mask,s=0.2,c='k')
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.title("$E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
#plt.xlim([-0,pi])
plt.show()

```

```python
import cmath 
cm = plt.cm.get_cmap('jet')
# plot with Rightmoving phase differce (arg(R1)-arg(R3))
eig_phase_R_mask1 = [(cmath.phase(x[0]) - cmath.phase(x[2])) for x in eigvec_array_mask1]
eig_phase_R_mask2 = [(cmath.phase(x[0]) - cmath.phase(x[2])) for x in eigvec_array_mask2]
eig_phase_R_mask3 = [(cmath.phase(x[0]) - cmath.phase(x[2])) for x in eigvec_array_mask3]
eig_phase_R_mask4 = [(cmath.phase(x[0]) - cmath.phase(x[2])) for x in eigvec_array_mask4]
Rphase_avg = np.mean(np.concatenate((eig_phase_R_mask1,eig_phase_R_mask2,eig_phase_R_mask3,eig_phase_R_mask4)))
plt.scatter(np.log(eig_array_mask1)/1j,E_array_mask,s=2,c=eig_phase_R_mask1, cmap = cm)
plt.scatter(np.log(eig_array_mask2)/1j,E_array_mask,s=2,c=eig_phase_R_mask2, cmap = cm)
plt.scatter(np.log(eig_array_mask3)/1j,E_array_mask,s=2,c=eig_phase_R_mask3, cmap = cm)
plt.scatter(np.log(eig_array_mask4)/1j,E_array_mask,s=2,c=eig_phase_R_mask4, cmap = cm)
plt.colorbar()
#plt.clim(0,pi)
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.title("Rphase average = "+str(round(Rphase_avg,2))+"\n $E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.show()

cm = plt.cm.get_cmap('jet')
# plot with leftmoving phase differce (arg(R1)-arg(R3))
eig_phase_L_mask1 = [(cmath.phase(x[1]) - cmath.phase(x[3])) for x in eigvec_array_mask1]
eig_phase_L_mask2 = [(cmath.phase(x[1]) - cmath.phase(x[3])) for x in eigvec_array_mask2]
eig_phase_L_mask3 = [(cmath.phase(x[1]) - cmath.phase(x[3])) for x in eigvec_array_mask3]
eig_phase_L_mask4 = [(cmath.phase(x[1]) - cmath.phase(x[3])) for x in eigvec_array_mask4]
Lphase_avg = np.mean(np.concatenate((eig_phase_L_mask1,eig_phase_L_mask2,eig_phase_L_mask3,eig_phase_L_mask4)))
plt.scatter(np.log(eig_array_mask1)/1j,E_array_mask,s=2,c=eig_phase_L_mask1, cmap = cm)
plt.scatter(np.log(eig_array_mask2)/1j,E_array_mask,s=2,c=eig_phase_L_mask2, cmap = cm)
plt.scatter(np.log(eig_array_mask3)/1j,E_array_mask,s=2,c=eig_phase_L_mask3, cmap = cm)
plt.scatter(np.log(eig_array_mask4)/1j,E_array_mask,s=2,c=eig_phase_L_mask4, cmap = cm)
plt.colorbar()
#plt.clim(0,pi)
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.title("Lphase average = "+str(round(Lphase_avg,2))+"\n $E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.show()
```

Input output states for superposition state

```python
# these all assume right going photon and no left going input states
# input states |psi>_in = r1m|k,1>_R + r3m|k-Delta,3>_R + l1p|k,1>_L + l3p|k-Delta,3>_L
# = AC|k,1>_R and BCexp(1j*(phi_a+phi_p))|k-Delta,3>_R
# outgoing states |psi>_out = r1p|k,1>_R + r3p|k-Delta,3>_R + l1m|k,1>_L + l3m|k-Delta,3>_L

def output_states(matrix,A,C,B,D,phi_a,phi_p,l1p,l3p,r1m,r3m):
    """
    takes in a 4 by 4 transfer matrix, input variables
    outputs the amplitude R1+
    """
    m11, m12, m13, m14 = matrix[0]
    m21, m22, m23, m24 = matrix[1]
    m31, m32, m33, m34 = matrix[2]
    m41, m42, m43, m44 = matrix[3]

    r1p = -((l3p*(m14*m22 - m12*m24))/(m24*m42 - m22*m44)) - (
      l1p*(-m14*m42 + m12*m44))/(
      m24*m42 - 
       m22*m44) - ((-m14*m22*m41 + m12*m24*m41 + m14*m21*m42 - 
           m11*m24*m42 - m12*m21*m44 + m11*m22*m44)*r1m)/(m24*m42 - 
         m22*m44) - ((m14*m23*m42 - m13*m24*m42 - m14*m22*m43 + 
           m12*m24*m43 + m13*m22*m44 - m12*m23*m44)*r3m)/(m24*m42 - 
         m22*m44)

    r3p = -((l3p*(-m24*m32 + m22*m34))/(m24*m42 - m22*m44)) - (
      l1p*(-m34*m42 + m32*m44))/(
      m24*m42 - 
       m22*m44) - ((m24*m32*m41 - m22*m34*m41 - m24*m31*m42 + 
           m21*m34*m42 + m22*m31*m44 - m21*m32*m44)*r1m)/(m24*m42 - 
         m22*m44) - ((-m24*m33*m42 + m23*m34*m42 + m24*m32*m43 - 
           m22*m34*m43 - m23*m32*m44 + m22*m33*m44)*r3m)/(m24*m42 - 
         m22*m44)

    l1m = (l3p*m24)/(m24*m42 - m22*m44) - (l1p*m44)/(
      m24*m42 - m22*m44) - ((m24*m41 - m21*m44)*r1m)/(
      m24*m42 - m22*m44) - ((m24*m43 - m23*m44)*r3m)/(m24*m42 - m22*m44)

    l3m = -((l3p*m22)/(m24*m42 - m22*m44)) + (l1p*m42)/(
      m24*m42 - m22*m44) - ((-m22*m41 + m21*m42)*r1m)/(
      m24*m42 - m22*m44) - ((m23*m42 - m22*m43)*r3m)/(m24*m42 - m22*m44)
    
    if m24*m42 - m22*m44 ==0:
        r1p, r3p, l1m, l3m = 0,0,0,0
        return r1p, r3p, l1m, l3m

    else:
        return r1p, r3p, l1m, l3m
```
