### Three-level Lambda Atom Coupled to a Waveguide


Import/ settings

```python
from cmath import exp
from math import pi
from matplotlib import pyplot as plt
from matplotlib import rc
import numpy as np
import time
rc('font',**{'family':'serif','serif':['Palatino']})
plt.rcParams.update({'font.size': 13})
rc('text', usetex=True)
```

Define reflection and transmission coefficients and transfer matrix

```python
def trans1(dtng21,vg,wW,g1,g3):
    """
    transmission of three-level atom through atom 1, with atom initially in 1
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |1>
    return 1 - (g1**2)/(1j*dtng21*(1-wW)*vg + g1**2 + g3**2)

def ref1(dtng21,vg,wW,g1,g3):
    """
    reflection of three-level atom through atom 1, with atom initially in 1
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |1>
    return - (g1**2)/(1j*dtng21*(1-wW)*vg + g1**2 + g3**2)

def trans3(dtng21,vg,wW,g1,g3):
    """
    transmission of three-level atom through atom 3, with atom initially in 1
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |1>
    return - (g1*g3)/(1j*dtng21*(1-wW)*vg + g1**2 + g3**2)

def ref3(dtng21,vg,wW,g1,g3):
    """
    reflection of three-level atom through atom 3, with atom initially in 1
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |1>
    return - (g1*g3)/(1j*dtng21*(1-wW)*vg + g1**2 + g3**2)

def trans1p(dtng21,vg,wW,g1,g3,deltaW):
    """
    transmission of three-level atom through atom 1, with atom initially in 3
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |3>
    # deltaW is dimensionaless Delta/detuning21
    return - (g1*g3)/(1j*dtng21*(1+deltaW-wW)*vg + g1**2 + g3**2)

def ref1p(dtng21,vg,wW,g1,g3,deltaW):
    """
    reflection of three-level atom through atom 1, with atom initially in 3
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |3>
    # deltaW is dimensionaless Delta/detuning21
    return - (g1*g3)/(1j*dtng21*(1+deltaW-wW)*vg + g1**2 + g3**2)

def trans3p(dtng21,vg,wW,g1,g3,deltaW):
    """
    transmission of three-level atom through atom 3, with atom initially in 3
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |3>
    # deltaW is dimensionaless Delta/detuning21
    return 1 - (g3**2)/(1j*dtng21*(1+deltaW-wW)*vg + g1**2 + g3**2)

def ref3p(dtng21,vg,wW,g1,g3,deltaW):
    """
    reflection of three-level atom through atom 3, with atom initially in 3
    """
    # wW is omega/detuning21
    # dtng is detuning
    # atom initially in |3>
    # deltaW is dimensionaless Delta/detuning21
    return - (g3**2)/(1j*dtng21*(1+deltaW-wW)*vg + g1**2 + g3**2)

def tmatrix(dtng21,vg,wW,g1,g3,deltaW):
    """
    transfer matrix of three level atom
    """
    t1 = trans1(dtng21,vg,wW,g1,g3)
    r1 = ref1(dtng21,vg,wW,g1,g3)
    t3 = trans3(dtng21,vg,wW,g1,g3)
    r3 = ref3(dtng21,vg,wW,g1,g3)
    t1p = trans1p(dtng21,vg,wW,g1,g3,deltaW)
    r1p = ref1p(dtng21,vg,wW,g1,g3,deltaW)
    t3p = trans3p(dtng21,vg,wW,g1,g3,deltaW)
    r3p = ref3p(dtng21,vg,wW,g1,g3,deltaW)
    # matrix elements
    M11 = -( (-r1p*r3*t1 + r1*r3*t1p + r1*r1p*t3 - t1*t1p*t3 - r1**2*t3p + t1**2*t3p)/(t1p*t3 - t1*t3p))
    M12 = -( (-r1p*t3 + r1*t3p)/(t1p*t3 - t1*t3p))
    M13 = -( (-r1p*r3p*t1 + r1*r3p*t1p + r1p**2*t3 - t1p**2*t3 - r1*r1p*t3p + t1*t1p*t3p)/(t1p*t3 - t1*t3p))
    M14 = -( (r1p*t1 - r1*t1p)/(t1p*t3 - t1*t3p))
    M21 = -( (r3*t1p - r1*t3p)/(t1p*t3 - t1*t3p))
    M22 = -( t3p/(t1p*t3 - t1*t3p))
    M23 = -( (r3p*t1p - r1p*t3p)/(t1p*t3 - t1*t3p))
    M24 =  t1p/(t1p*t3 - t1*t3p)
    M31 = -( (-r3*r3p*t1 + r3**2*t1p + r1*r3p*t3 - t1p*t3**2 - r1*r3*t3p + t1*t3*t3p)/(t1p*t3 - t1*t3p))
    M32 = -( (-r3p*t3 + r3*t3p)/(t1p*t3 - t1*t3p))
    M33 = -( (-r3p**2*t1 + r3*r3p*t1p + r1p*r3p*t3 - r1p*r3*t3p - t1p*t3*t3p + t1*t3p**2)/(t1p*t3 - t1*t3p))
    M34 = -( (r3p*t1 - r3*t1p)/(t1p*t3 - t1*t3p))
    M41 = -((-r3*t1 + r1*t3)/(t1p*t3 - t1*t3p))
    M42 = +( t3/(t1p*t3 - t1*t3p))
    M43 = -( (-r3p*t1 + r1p*t3)/(t1p*t3 - t1*t3p))
    M44 = -( t1/(t1p*t3 - t1*t3p))
    return np.array([[M11,M12,M13,M14],[M21,M22,M23,M24], [M31,M32,M33,M34],[M41,M42,M43,M44]])
```

Plot transmission and reflection from transfer matrix

```python
def tandr_fromtmatrix(dtng21,vg,wW,g1,g3,deltaW):
    """
    transfer matrix of three level atom
    """
    t1 = trans1(dtng21,vg,wW,g1,g3)
    r1 = ref1(dtng21,vg,wW,g1,g3)
    t3 = trans3(dtng21,vg,wW,g1,g3)
    r3 = ref3(dtng21,vg,wW,g1,g3)
    t1p = trans1p(dtng21,vg,wW,g1,g3,deltaW)
    r1p = ref1p(dtng21,vg,wW,g1,g3,deltaW)
    t3p = trans3p(dtng21,vg,wW,g1,g3,deltaW)
    r3p = ref3p(dtng21,vg,wW,g1,g3,deltaW)
    # matrix elements
    M11 = -( (-r1p*r3*t1 + r1*r3*t1p + r1*r1p*t3 - t1*t1p*t3 - r1**2*t3p + t1**2*t3p)/(t1p*t3 - t1*t3p))
    M12 = -( (-r1p*t3 + r1*t3p)/(t1p*t3 - t1*t3p))
    M13 = -( (-r1p*r3p*t1 + r1*r3p*t1p + r1p**2*t3 - t1p**2*t3 - r1*r1p*t3p + t1*t1p*t3p)/(t1p*t3 - t1*t3p))
    M14 = -( (r1p*t1 - r1*t1p)/(t1p*t3 - t1*t3p))
    M21 = -( (r3*t1p - r1*t3p)/(t1p*t3 - t1*t3p))
    M22 = -( t3p/(t1p*t3 - t1*t3p))
    M23 = -( (r3p*t1p - r1p*t3p)/(t1p*t3 - t1*t3p))
    M24 =  t1p/(t1p*t3 - t1*t3p)
    M31 = -( (-r3*r3p*t1 + r3**2*t1p + r1*r3p*t3 - t1p*t3**2 - r1*r3*t3p + t1*t3*t3p)/(t1p*t3 - t1*t3p))
    M32 = -( (-r3p*t3 + r3*t3p)/(t1p*t3 - t1*t3p))
    M33 = -( (-r3p**2*t1 + r3*r3p*t1p + r1p*r3p*t3 - r1p*r3*t3p - t1p*t3*t3p + t1*t3p**2)/(t1p*t3 - t1*t3p))
    M34 = -( (r3p*t1 - r3*t1p)/(t1p*t3 - t1*t3p))
    M41 = -((-r3*t1 + r1*t3)/(t1p*t3 - t1*t3p))
    M42 = +( t3/(t1p*t3 - t1*t3p))
    M43 = -( (-r3p*t1 + r1p*t3)/(t1p*t3 - t1*t3p))
    M44 = -( t1/(t1p*t3 - t1*t3p))
    t1 = -M44/(M24*M42-M22*M44) #S22
    r1 = (M24*M41-M21*M44)/(M24*M42-M22*M44) #S21
    t3 = -M22/(M24*M42-M22*M44) #S44
    r3 = -(M23*M42-M22*M43)/(M24*M42-M22*M44) #S43
    t1p = M24/(M24*M42-M22*M44) #S24
    r1p = -(M24*M43-M23*M44)/(M24*M42-M22*M44) #S23
    t3p = M42/(M24*M42-M22*M44) #S42
    r3p = -(M21*M42-M22*M41)/(M24*M42-M22*M44) #S21
    return [t1,r1,t3,r3,t1p,r1p,t3p,r3p]

t1_array = []
r1_array = []
t3_array = []
r3_array = []
t1p_array = []
r1p_array = []
t3p_array = []
r3p_array = []
for wW in wW_array:
    t1,r1,t3,r3,t1p,r1p,t3p,r3p = tandr_fromtmatrix(dtng21,vg,wW,g1,g3,deltaW)
    t1_array.append(t1)
    r1_array.append(r1)
    t3_array.append(t3)
    r3_array.append(r3)
    t1p_array.append(t1p)
    r1p_array.append(r1p)
    t3p_array.append(t3p)
    r3p_array.append(r3p)
# plot from transfer matrix
plt.plot(wW_array,[abs(x)**2 for x in t1_array], label= '$|t_1|^2$')
plt.plot(wW_array,[abs(x)**2 for x in r1_array], label= '$|r_1|^2$')
plt.plot(wW_array,[abs(x)**2 for x in t3_array], label= '$|t_3|^2$')
plt.plot(wW_array,[abs(x)**2 for x in r3_array], label= '$|r_3|^2$')
plt.plot(wW_array,[abs(x)**2 for x in t1p_array], label= '$|t_1\prime|^2$')
plt.plot(wW_array,[abs(x)**2 for x in r1p_array], label= '$|r_1\prime|^2$')
plt.plot(wW_array,[abs(x)**2 for x in t3p_array], label= '$|t_3\prime|^2$')
plt.plot(wW_array,[abs(x)**2 for x in r3p_array], label= '$|r_3\prime|^2$')
plt.xlabel("$\omega/\Omega$")
plt.legend()
plt.show()
```

Plot dispersion relation

```python
# Define variables
dtng21 = 20
vg = 1
g1 = 1
g3 = 1
dl = 3.4 * 10**(-1)
deltaW = 0
# Define plot range
ww_min = 0
ww_max = pi
steps = 4000
wW_array = np.linspace(ww_min, ww_max, steps)

eig_array1 = []
eig_array2 = []
eig_array3 = []
eig_array4 = []

for wW in wW_array:
    Mprop = np.array([[exp(1j*2*pi*wW*dl),0,0,0],\
                       [0,exp(-1j*2*pi*wW*dl),0,0],\
                       [0,0,exp(1j*2*pi*(wW-deltaW)*dl),0],\
                       [0,0,0,exp(-1j*2*pi*(wW-deltaW)*dl)]])
    Q = np.dot(Mprop, tmatrix(dtng21,vg,wW,g1,g3,deltaW))
    #print(round(np.linalg.det(tmatrix(dtng21,vg,wW,g1,g3,deltaW)),2)) -> check determinant is 1
    eig = np.linalg.eigvals(Q)
    eig_array1.append(eig[0])
    eig_array2.append(eig[1])
    eig_array3.append(eig[2])
    eig_array4.append(eig[3])

    
plt.scatter(np.log(eig_array1)/1j,wW_array,s=0.2,c='k')
plt.scatter(np.log(eig_array2)/1j,wW_array,s=0.2,c='k')
plt.scatter(np.log(eig_array3)/1j,wW_array,s=0.2,c='k')
plt.scatter(np.log(eig_array4)/1j,wW_array,s=0.2,c='k')
plt.ylabel("$\omega/\Omega$")
plt.xlabel("$K d$")
plt.xlim([-0,pi])
plt.show()
```

dispersion relation

```python
def tmatrix(E2,E,vg,g1,g3):
    """
    transfer matrix of three level atom
    """
    t1 = trans1(E2,E,vg,g1,g3)
    r1 = ref1(E2,E,vg,g1,g3)
    t3 = trans3(E2,E,vg,g1,g3)
    r3 = ref3(E2,E,vg,g1,g3)
    t1p = trans1p(E2,E,vg,g1,g3)
    r1p = ref1p(E2,E,vg,g1,g3)
    t3p = trans3p(E2,E,vg,g1,g3)
    r3p = ref3p(E2,E,vg,g1,g3)
    # matrix elements
    denominator = (t1p*t3 - t1*t3p)
    if (t1p*t3 - t1*t3p) == 0:
        denominator = 0.00000001 #avoid divide by zero error
    M11 = -( (-r1p*r3*t1 + r1*r3*t1p + r1*r1p*t3 - t1*t1p*t3 - r1**2*t3p + t1**2*t3p)/denominator)
    M12 = -( (-r1p*t3 + r1*t3p)/denominator)
    M13 = -( (-r1p*r3p*t1 + r1*r3p*t1p + r1p**2*t3 - t1p**2*t3 - r1*r1p*t3p + t1*t1p*t3p)/denominator)
    M14 = -( (r1p*t1 - r1*t1p)/denominator)
    M21 = -( (r3*t1p - r1*t3p)/denominator)
    M22 = -( t3p/denominator)
    M23 = -( (r3p*t1p - r1p*t3p)/denominator)
    M24 =  t1p/denominator
    M31 = -( (-r3*r3p*t1 + r3**2*t1p + r1*r3p*t3 - t1p*t3**2 - r1*r3*t3p + t1*t3*t3p)/denominator)
    M32 = -( (-r3p*t3 + r3*t3p)/denominator)
    M33 = -( (-r3p**2*t1 + r3*r3p*t1p + r1p*r3p*t3 - r1p*r3*t3p - t1p*t3*t3p + t1*t3p**2)/denominator)
    M34 = -( (r3p*t1 - r3*t1p)/denominator)
    M41 = -((-r3*t1 + r1*t3)/denominator)
    M42 = +( t3/denominator)
    M43 = -( (-r3p*t1 + r1p*t3)/denominator)
    M44 = -( t1/denominator)
    return np.array([[M11,M12,M13,M14],[M21,M22,M23,M24], [M31,M32,M33,M34],[M41,M42,M43,M44]])
```

```python
# Define variables
E2=20
E1=0
E3=10
vg = 1
g1 = 1
g3 = 1
llambda = 2*pi*vg/(E2-E1)
d = 3.4 * 10**(-1) *llambda

# Define plot range
E_min = 0
E_max = 42
steps = 4000
E_array = np.linspace(E_min, E_max, steps)

eig_array1 = []
eig_array2 = []
eig_array3 = []
eig_array4 = []

for E in E_array:
    Mprop = np.array([[exp(1j*(E-E1)*d/vg),0,0,0],\
                       [0,exp(-1j*(E-E1)*d/vg),0,0],\
                       [0,0,exp(1j*(E-E3)*d/vg),0],\
                       [0,0,0,exp(-1j*(E-E3)*d/vg)]])
    Q = np.dot(Mprop, tmatrix(E2,E,vg,g1,g3))
    eig = np.linalg.eigvals(Q)
    #print(np.around(eig,2)) #-> check determinant is 1
    eig_array1.append(eig[0])
    eig_array2.append(eig[1])
    eig_array3.append(eig[2])
    eig_array4.append(eig[3])

plt.scatter([x.real for x in eig_array1],[x.imag for x in eig_array1])
plt.ylabel("$Im(Kd)$")
plt.xlabel("$Re(Kd)$")
plt.show()
    
plt.scatter(np.log(eig_array1)/1j,E_array,s=0.2,c='k')
plt.scatter(np.log(eig_array2)/1j,E_array,s=0.2,c='k')
plt.scatter(np.log(eig_array3)/1j,E_array,s=0.2,c='k')
plt.scatter(np.log(eig_array4)/1j,E_array,s=0.2,c='k')
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.title("$E_2 = $"+str(E2)+"$, E_1 = $"+str(E1)+"$, E_3 = $"+str(E3))
plt.xlim([-0,pi])
plt.show()
```

## Fabry Perot

first do 2level atom

```python
E1 = 20
vg = -1
g = 1
dl = 3.4 * 10**(-3)

def t_twolvl(E1,vg,E,g):
    """
    transmission of two-level atom
    """
    # wW is omega/detuning
    # dtng is detuning
    return (E-E1)*vg / ((1j*g**2+ (E-E1)*vg))

def r_twolvl(E1,vg,E,g):
    """
    reflection of two-level atom
    """
    return 1j*g**2 / ((1j*g**2+ (E-E1)*vg))

def tmatrix_twolvl(E1,vg,E,g):
    """
    transfer matrix of two level atom
    """
    t = t_twolvl(E1,vg,E,g)
    r = r_twolvl(E1,vg,E,g)
    # matrix elements
    if t ==0:
        t=0.0000001 #avoid divide by zero
    M11 = (t**2-r**2)/t
    M12 = r/t
    M21 = -r/t
    M22 = 1/t
    return np.array([[M11,M12],[M21,M22]])

# PLOT TRANSMISSION AND REFLECTION ####################################
# Define plot range
E_min = 0
E_max = 40
steps = 1000
E_array = np.linspace(E_min, E_max, steps)

t_array = []
r_array = []
for E in E_array:
    matrix = tmatrix_twolvl(E1,vg,E,g)
    r_array.append(-matrix[1][0]/matrix[1][1])
    t_array.append(-1/matrix[1][1])
    
# plot from transfer matrix
plt.plot(E_array,[abs(x)**2 for x in t_array], label= '$|t|^2$')
plt.plot(E_array,[abs(x)**2 for x in r_array], label= '$|r|^2$')
plt.xlabel("$E$")
plt.legend()
plt.show()

# compare to analytical
plt.plot(E_array,[abs(t_twolvl(dtng,vg,E,g))**2 for E in E_array], label= '$|t|^2$')
plt.plot(E_array,[abs(r_twolvl(dtng,vg,E,g))**2 for E in E_array], label= '$|t|^2$')
plt.xlabel("$E$")
plt.legend()
plt.show()
```

dispersion of two-level

```python
# PLOT DISPERSION RELATION ####################################
# Define variables
E1 = 20
vg = 1
g = 1
llambda = 2*pi*vg/(E1)
d = 3.4 * 10**(-1) *llambda

# Define plot range
E_min = 0
E_max = 120
steps = 4000
E_array = np.linspace(E_min, E_max, steps)

eig_array1 = []
eig_array2 = []
for E in E_array:
    Mprop = np.array([[exp(1j*(E-E1)*d/vg),0],[0,exp(-1j*(E-E1)*d/vg)]])
    Q = np.dot(Mprop, tmatrix_twolvl(E1,vg,E,g))
    eig = np.linalg.eigvals(Q)
    eig_array1.append(eig[0])
    eig_array2.append(eig[1])
    
plt.scatter(np.log(eig_array1)/1j,E_array,s=0.2)
plt.scatter(np.log(eig_array2)/1j,E_array,s=0.2)
plt.ylabel("$E$")
plt.xlabel("$K d$")
plt.xlim([-0,pi])
plt.show()
```
